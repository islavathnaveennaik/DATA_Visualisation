#applying the different formatting lines

```python
#format of line---> dotted, solid, dashed, dash dot
import numpy as np
x = np.array([1,2,3,4,6,8])
y = np.array([3,8,1,5,10,7])
plt.plot(x,y,'o:r')
plt.show()
```

```python
plt.plot(x,y,'o-r')   #solid line
plt.show()
```

```python
plt.plot(x,y,'o--r') #dashed lines
plt.show()
```

```python
plt.plot(x,y,'o-.r') #dash dot lines
plt.show()
```

```python
#change size of the marker
#use the parameter "Marker size-->ms"
plt.plot(x,y,'o:g',ms=10)
plt.show()
```

```python
#give the boundary colur to the marker--> marker edge color-->mec
plt.plot(x,y,'p-.y',ms=10,mec='b')
plt.show()
```

```python
#fill the marker color with different color-->use marker fill color--> mfc
plt.plot(x,y,'P-.g',ms=10,mec='b',mfc= '#2ac2c7')
plt.show()
```

#Encodings


```python
#1.Label encoding:- converts the ach catgory in to unique integers
from sklearn.preprocessing import LabelEncoder
fruits = ['apple','banana','orange','pineapple','kiwi','watermelon','apple']
label_encode = LabelEncoder()
encoded_fruits = label_encode.fit_transform(fruits)
print(encoded_fruits)
#decode the encoded part
orginal = label_encode.classes_
#if you want to get the particular encoded item to be decoded use "orginal = label_encode.classes_[encode_number]"
print(orginal)
```

```python
#one-hot encoding:- create the binary columns for category
import pandas as pd
```

```python
data = {
    'fruits':['apple','banana','orange','pineapple','kiwi','watermelon','apple']
}
#convert the bove data in to dataframe
df = pd.DataFrame(data)
df_encode = pd.get_dummies(df,columns=['fruits'])
print(df_encode)
colums = df_encode.columns
print(colums)
#pandas.getdummies will create binaray columns for each unique category
```
